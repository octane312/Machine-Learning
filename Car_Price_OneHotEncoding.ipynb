# **ONE HOT ENCODING**

## **IMPORTING MODULES**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.compose import ColumnTransformer
```

## **LOADING DATASET**

```python
d=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/5_one_hot_encoding/Exercise/carprices.csv")
d
```

Encoding the Car Model

```python
le=LabelEncoder()
d["Car Model"]=le.fit_transform(d["Car Model"])
print(d)
```

## **SEPERATING DATA SET**

```python
x=d[["Car Model","Mileage","Age(yrs)"]]
Y=d["Sell Price($)"]
```

Applying One Hot Encoding

```python
ct = ColumnTransformer([('Car Model', OneHotEncoder(),[0])], remainder='passthrough')
X = ct.fit_transform(x)
X=X[:,1:]
print(X)
```

## **FITTING MODEL**

```python
model=LinearRegression()
model.fit(X,Y)
```

## **PREDICTING USING MODEL**

Here I take two test case

1.   Mercedes Benz that is 4 years old with mileage 45000
2.   BMW X5 that is 7 year old with mileage 86000

```python
Y_pred=model.predict(X)
print(f"Test case 1 : {model.predict([[0,1,45000,4]])}\nTest case 1 : {model.predict([[1,0,86000,7]])}")
```

## **EVALUATING MODEL**

```python
print(f"Accuracy Score : {model.score(X,Y_pred)}")
```
