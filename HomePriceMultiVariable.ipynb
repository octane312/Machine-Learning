# **HOME PRICE PREDICTION WITH MULTI_VARIABLE**

## **IMPORTING MODULES**

```python
import pandas as pd
import math as m
from sklearn.linear_model import LinearRegression
```

## **LOADING DATASET**

```python
data=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/2_linear_reg_multivariate/homeprices.csv")
Y=data.price
X=data.drop("price",axis=1)
X
```

Since number of bedrooms is not available on index 2, we will fill it with median of number of bedrooms.

```python
med=m.floor(X.bedrooms.median())
X.fillna({"bedrooms":med},inplace=True)
X
```

## **FITTING MODEL**

```python
model=LinearRegression()
model.fit(X,Y)
```

## **COEFFICIENTS AND INTERCEPT OF MODEL**

```python
print(model.coef_)
print(model.intercept_)
```

## **PREDICTING VALUE**

Here I am taking two test cases.
i) 3000 sqft area 3 bedrooms, 40 year old
ii) 2500 sqft area 4 bedrooms, 5 year old

### **CREATING A DATASET WITH TEST CASES**

```python
test=pd.DataFrame([[3000,3,40],[2500,4,5]],columns=["area","bedrooms","age"])
test
```

### **PREDICTING TEST VALUES**

```python
print(f"Test Case 1 : {model.predict(test)[0]} \nTest Case 2 : {model.predict(test)[1]}")
```
